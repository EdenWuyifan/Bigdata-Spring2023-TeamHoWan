```python
import datamart
import datamart_rest
```

```python
client = datamart_rest.RESTDatamart('https://auctus.vida-nyu.org/api/v1')
```

```python
query = datamart.DatamartQuery(
    keywords=['Daily Ozone Emissions'],
    variables=[]
)
cursor = client.search(query=query)
```

```python
page_data = cursor.get_next_page()
single_data = page_data[0].get_json_metadata()
```
